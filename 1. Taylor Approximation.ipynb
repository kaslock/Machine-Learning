```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Define a differentiable function that maps from real number to real number
# Define a domain of the function.

x = np.linspace(0, 5, 500)
y = (x+1)*(x-2)**2
```

```python
# Plot the function

plt.plot(x, y, "BLUE")
plt.title('Graph')
plt.xlabel('X AXIS')
plt.ylabel('Y AXIS')
plt.show()
```

**Select a point within the domain**

$Point(a, y(a))$

**$a = 3$　　$Point(3, 4)$**

```python
 # Mark the selected point on the function
    
plt.plot(x, y, "BLUE")
plt.scatter(3, 4, c = 'g')
plt.title('Point')
plt.xlabel('X AXIS')
plt.ylabel('Y AXIS')
plt.show()
```

$$First-order　Taylor　approximation$$

$$f(x) \approx f(a) + f^\prime(a)(x-a)$$

**Define the first-order Taylor approximation at the selected point**

$f(x) \approx f(3) + f^\prime(3)(x-3)$

$f(3) = 4$  $f^\prime(3) = 9$

$\hat{y} = 9x-23$

```python
T = 9*x - 23
```

```python
# Plot the Taylor approximation with the same domain of the original function.

plt.plot(x, T, "RED")
plt.plot(x, y, "BLUE")
plt.scatter(3, 4, c = 'g')
plt.title('Taylor Approximation')
plt.xlabel('X AXIS')
plt.ylabel('Y AXIS')
plt.show()
```
