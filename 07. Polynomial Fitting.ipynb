**Solve a least square problem to find an optimal polynomial curve for a given set of two dimensional points.**

**Demonstrate the effect of the degree of polynomial in fitting a given set of points.**

```python
import matplotlib.pyplot as plt
import numpy as np
from numpy.linalg import *
```

```python
#
# example of distance function between two vectors x and y
#
def distance(x, y):

    d = (x - y) ** 2
    s = np.sum(d)
    r = np.sqrt(s)

    return(r)

# Define a domain of the function.
x_lim = 10

# Define Function
x = np.linspace(-x_lim,x_lim)
y = x**3 -2*x**2 + 3*x + 1

#
# Plot the function
#
plt.plot(x, y, "BLUE")
plt.title('Graph')
plt.xlabel('X AXIS')
plt.ylabel('Y AXIS')
plt.show()
```

- choose a polynomial curve and generate points along the curve with random noise

- plot the generated noisy points along with its original polynomial without noise

```python
x_cnt = 200
b = np.empty((x_cnt, 1), dtype=float)

# Get Noise
noise = np.random.randint(-200, 200, x_cnt)
cnt = 0

# Plot the function
x = np.linspace(-x_lim, x_lim)
y = x**3 -2*x**2 + 3*x + 1
plt.plot(x, y, "BLUE")

# Plot the noise
for x in np.arange(-x_lim, x_lim, 0.1):
    b[cnt] = y = x**3 -2*x**2 + 3*x + 1 + noise[cnt]
    cnt += 1
    plt.scatter(x, y, c='k', s=3)

plt.title('Graph and Noise Point')
plt.xlabel('X AXIS')
plt.ylabel('Y AXIS')
plt.show()
```

```python
def Approximating_Polynomial_Curve(degree):
    A = np.empty((x_cnt, degree+1), dtype=float)
    b = np.empty((x_cnt, 1), dtype=int)
    
    cnt = 0
    for x in np.arange(-x_lim, x_lim, 0.1):
        A[cnt, 0] = 1
        for n in range(degree):
            A[cnt, n+1] = x**(n+1)
        cnt += 1
    
    cnt = 0
    for i in np.arange(-x_lim, x_lim, 0.1):
        b[cnt] = y = i**3 -2*i**2 + 3*i + 1 + noise[cnt]
        cnt += 1
        
    A = np.asmatrix(A)
    b = np.asmatrix(b)

    theta = inv(A.T*A)*A.T*b
    
    X = np.asfarray(A*theta)
    Y = np.asfarray(b)
    
    # Get energy with L2_Norm
    energy_list[degree] = distance(X, Y)
    th = np.asfarray(theta)
    
    # Plot the noise
    cnt = 0
    for x in np.arange(-x_lim, x_lim, 0.1):
        y = x**3 -2*x**2 + 3*x + 1 + noise[cnt]    
        plt.scatter(x, y, c='k', s=3)
        cnt += 1
    
    
    return np.array(th)
```

- plot the approximating polynomial curve obtained by solving a least square problem

# . $f_\theta(x) = \theta_0 + \theta_1 x$

```python
degree = 1
th = Approximating_Polynomial_Curve(degree)

print('**theta')
print(th)

x = np.linspace(-x_lim, x_lim)
y = th[1]*x + th[0]
plt.plot(x, y, 'c')
plt.title('Degree = 1')
plt.show()
```

# . $f_\theta(x) = \theta_0 + \theta_1 x + \theta_2 x^2$

```python
degree = 2
th = Approximating_Polynomial_Curve(degree)

print('**theta')
print(th)

x = np.linspace(-x_lim, x_lim)
y = th[2]*x**2 + th[1]*x + th[0]
plt.plot(x, y, 'g')
plt.title('Degree = 2')
plt.show()
```

```python
degree = 3
th = Approximating_Polynomial_Curve(degree)

print('**theta')
print(th)

x = np.linspace(-x_lim, x_lim)
y = th[3]*x**3 + th[2]*x**2 + th[1]*x + th[0]
plt.plot(x, y, 'y')
plt.title('Degree = 3')
plt.show()
```
